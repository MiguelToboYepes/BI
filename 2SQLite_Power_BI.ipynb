```python
#Usualmente encontramos informaciones de las organizaciones almacenada de forma aislada.

#Cargamos las librerias necesarias

import pandas as pd
import sqlite3

#Le damos lectura a los archivos de excel

datos_productos= pd.read_csv("/info_productos.csv")
datos_ventas= pd.read_csv("/info_ventas.csv")
```

```python
datos_productos.head()
```

```python
datos_ventas.head()
```

```python
print("Dimension tabla productos",datos_productos.shape)
print("Dimension tabla ventas",datos_ventas.shape)
```

```python
#Se hace la conexión a la base de datos. Si NO existe se crea
conn=sqlite3.connect("veaycompre.db")
c=conn.cursor()
```

```python
#Creamos la tabla producto
c.execute('''CREATE TABLE productos
(id_producto INTEGER PRIMARY KEY,
nombre TEXT,
precio REAL)''')
```

```python
#Pasamos la informacion de DataFrame a la tabla de SQLite

datos_productos.to_sql("productos",conn, if_exists="append",index=False)
```

```python
#Agregar informacion a la tabla de productos directamente

#Hacemos una lista con la informacion faltante
productos_faltantes=[(10001,"Laptop",1200.00),
                     (10002,"Coco",2.00),
                     (10003,"Cuaderno",1.50)]

#insertamos la informacion en una tabla exiteste
c.executemany("INSERT INTO productos VALUES(?,?,?)",productos_faltantes)
```

```python
#Creamos la tabla de ventas
c.execute('''CREATE TABLE ventas
(id_ventas INTEGER PRIMARY KEY,
id_producto INTEGER,
cantidad INTEGER,
fecha DATE,
FOREIGN KEY(id_producto) REFERENCES productos(id_producto))''')
```

```python
#Pasar todo del DataFrame a la nueva tabla
datos_ventas.to_sql("ventas",conn, if_exists="append",index=False)
#Guardamos los cambios
conn.commit()
#cerranos la conexion
conn.close()
```
